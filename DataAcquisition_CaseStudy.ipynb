5. For a numerical column of your choice, calculate the mean, median, and
standard deviation.

```python
mean = df['Study Hours per Week'].mean()
print(f"Mean Value for Study Hours Per Week = {mean}")

median = df['Study Hours per Week'].median()
print(f"Median Value for Study Hours Per Week = {median}")

std_dev = df['Study Hours per Week'].std()
print(f"Standard Deviation for Study Hours Per Week = {std_dev}")
```

6. Identify any potential outliers in a numerical column of your choice. Explain
your approach.

```python
column = 'Study Hours per Week'
Q1 = df[column].quantile(0.25)
Q3 = df[column].quantile(0.75)
IQR = Q3-Q1
lower_bound = Q1-1.5*IQR
upper_bound = Q3+1.5*IQR
outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
outliers
```

8. Group the data by a categorical column and compute the aggregate statistics
(e.g., mean, sum) for a numerical column.

```python
aggregate_stats = df.groupby('Participation in Extracurricular Activities').agg({
    'Study Hours per Week': ['mean', 'sum'],
    'Attendance Rate': ['mean', 'sum'],
    'Previous Grades': ['mean', 'sum']
})

print(aggregate_stats)
```
